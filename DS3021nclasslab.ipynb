## Quesiton 1: What are the COVID-19 mortality rates by age group? Which age groups are most at risk?

## Pseudocode
1. Load the COVID-19 dataset.
2. Clean missing data
3. Group data by age group
4. Sort by mortality rate in descending order.
5. Display data table.


## Step 1: Import Data

```python
import pandas as pd 
df = pd.read_csv("COVID-19_final.csv")
df.head()
```

## Step 2: Clean Missing Data

```python
df.isna().sum()
df = df.dropna(subset=['age'])
df['age'] = df['age'].astype(float)
```

## Step 3: Group Data by Age

```python
def agefunction(age):
    if age < 21:
        return "0-20"
    elif age < 31:
        return "20-30"
    elif age < 41:
        return "30-40"
    elif age < 51:
        return "40-50"
    elif age < 61:
        return "50-60"
    elif age < 71:
        return "60-70"
    elif age < 81:
        return "70-80"
    else :
        return "80+"
    
df['agegroup'] = df['age'].apply(agefunction)
df.head()
```

## Step 4: Build a New Datafram with Total Patient and Total Deceased

```python
df2 = df.groupby("agegroup").agg(
        totalnum = ('age', 'count'),
        deceased = ('outcome', lambda x: (x == "Deceased").sum()))

df2.head()
```

## Step 5: Calculate the Death Rate for Different Group of Patient

```python
df2['deathrate'] = df2["deceased"] / df2["totalnum"]

df2.head()
```

## Step 6: Build the Function

```python
import pandas as pd

def death_rate(df):
    df['agegroup'] = df['age'].apply(agefunction) 
    grouped = df.groupby('agegroup')
    Result = pd.DataFrame({
        'totalnum': grouped['age'].count(),  
        'deceased': grouped['outcome'].apply(lambda x: (x == 'Deceased').sum()),  
    })
    Result['death_rate'] = Result['deceased'] / Result['totalnum']
    
    return Result
df_test = death_rate(df)
df_test.head()


```

## Step 7: Conclusion and Results

As we see in the dataset, the deathrate decrease when the age increase, so it said there is a positive relationsip between age and deathrate, but we do not know the causation in this case.
